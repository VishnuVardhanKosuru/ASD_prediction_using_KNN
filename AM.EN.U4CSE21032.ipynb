```python
# Create separate scaler and label encoder instances for each dataset
scaler_child = StandardScaler()
scaler_teen = StandardScaler()
scaler_adult = StandardScaler()

label_child = LabelEncoder()
label_teen = LabelEncoder()
label_adult = LabelEncoder()
```

```python
numerical_cols = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10_Autism_Spectrum_Quotient', 'Age_Years']
categorical_cols = ['Sex', 'Ethnicity', 'Jaundice', 'Family_mem_with_ASD', 'Who_completed_the_test']
```

```python
X_child_std = pd.DataFrame(scaler_child.fit_transform(X_child[numerical_cols]), columns=numerical_cols, index=X_child.index)
for col in categorical_cols:
    X_child_std[col] = label_child.fit_transform(X_child[col])
```

```python
X_adult_std = pd.DataFrame(scaler_adult.fit_transform(X_adult[numerical_cols]), columns=numerical_cols, index=X_adult.index)
for col in categorical_cols:
    X_adult_std[col] = label_adult.fit_transform(X_adult[col])
```

```python
X_teen_std = pd.DataFrame(scaler_teen.fit_transform(X_teen[numerical_cols]), columns=numerical_cols, index=X_teen.index)
for col in categorical_cols:
    X_teen_std[col] = label_teen.fit_transform(X_teen[col])
```
